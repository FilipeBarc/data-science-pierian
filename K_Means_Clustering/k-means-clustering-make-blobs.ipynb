```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
%matplotlib inline
sns.set_style('darkgrid')
```

### Importing data

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
data[0].shape
```

### Ploting the data

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow',edgecolors='black')
```

### Creating a kmeans clustering algorithm

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4,n_init='auto')
```

```python
kmeans.fit(data[0])
```

### Getting centers and labels

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

### Predict labels compared to actual labels 

```python
fig , (ax1,ax2) = plt.subplots(1,2, sharey=True,figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow',edgecolors='black')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow',edgecolors='black')
```
