```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Loading built-in data set

```python
tips = sns.load_dataset('tips')
tips
```

### Bar plot - aggregate categorical data based in some function

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

### Count plot - count the number of occurrences 

```python
sns.countplot(x='sex',data=tips)
```

### Box plot - show the distribution of categorical data

```python
sns.boxplot(x='day',y='total_bill',data=tips,showmeans=True,meanprops={"marker":"o",
                       "markerfacecolor":"white", 
                       "markeredgecolor":"black",
                      "markersize":"3"})
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

### Box plot - showing the average

```python
plt.figure(figsize=(8,8))
ax = sns.boxplot(x='day',y='total_bill',data=tips)

mean = tips.groupby(['day'])['total_bill'].mean()
vert = tips['total_bill'].mean() * 0
cat = ax.get_xticks()

for x in cat:
    ax.text(x,mean[x] + vert,np.round(mean[x],2),
            horizontalalignment='center',size='small',color='w',weight='semibold')
```

### Box plot - showing the median

```python
ax = sns.boxplot(x='day',y='total_bill',data=tips)

med = tips.groupby(['day'])['total_bill'].median()
vert = tips['total_bill'].median() * 0.05 
cat = ax.get_xticks()

for x in cat:
    ax.text(x,med[x] + vert,med[x],horizontalalignment='center',size='small',color='w',weight='semibold')
```

### Violin plot - show the distribution of categorical data

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

### Strip plot - draw a scatter plot where one variable is categorical

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex')
```

### Swarm plot - similar to strip plot, but points dont overlap

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

### Violin plot with Swarm plot

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

### Cat plot - can be any categorical plot by changing the kind

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='box')
```
