### Dropping data  and zip code columns

```python
df = df.drop('date',axis=1)
```

```python
df = df.drop('zipcode',axis=1)
```

### Train, test, split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('price',axis=1).values
y = df['price'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

### Scaling/normalizing

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

### Creating the model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

### Training the model

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),batch_size=128,epochs=130)
```

### Checking the losses and validation losses

```python
loss = pd.DataFrame(model.history.history)
```

```python
loss.plot()
```

### Predicting and checking the erros

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```python
pred = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,pred))
```

```python
mean_absolute_error(y_test,pred)
```

### Checking the mean abs error compared to the mean price 

```python
print('\n')
print('The mean abs error is', f'{mean_absolute_error(y_test,pred)/df["price"].mean()*100:.2f}','% of the price mean value, thats not good')
```

### Using the explained variance score to get a look deeper understanding of the evaluation metrics

```python
explained_variance_score(y_test,pred)
```
