```python
import seaborn as sns
%matplotlib inline
```

### Built-in data sets 

```python
tips = sns.load_dataset('tips')
tips
```

### Dis plot - single variable

```python
sns.displot(tips['total_bill'],kde=True,bins=20)
```

### Joint plot - combined variables

```python
sns.jointplot(x='total_bill',y='tip',data=tips,s=100,color='red')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

### Pair plot - pair relationships

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

### Rug plot - single variables represented with dashes

```python
sns.rugplot(tips['total_bill'])
```

### KDE plot - single variables represented with a line

```python
sns.kdeplot(tips['total_bill'])
```
