```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Importing data

```python
df1 = pd.read_csv('df1',index_col=0)
df1
```

```python
df2 = pd.read_csv('df2')
df2
```

### Histogram plot

```python
df1['A'].hist(bins=30,lw=1,edgecolor='black')
```

```python
df1['A'].plot(kind='hist',edgecolor='black')
```

### Area plot

```python
df2.plot.area(alpha=0.4)
```

### Bar plot

```python
df2.plot.bar(alpha=0.5)
```

```python
df2.plot.bar(alpha=0.5,stacked=True)
plt.legend(loc='center left',bbox_to_anchor=(1.0,0.5))
```

### Line plot

```python
df1['B'].plot.line(figsize=(10,3))
```

### Scatter plot

```python
df1.plot.scatter(x='A',y='B',s=50,edgecolor='black',figsize=(12,3),c='red')
```

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*100,edgecolor='black')
```

### Box plots

```python
df2.plot.box(showmeans=True,meanprops={"marker":"o",
                       "markerfacecolor":"white", 
                       "markeredgecolor":"black",
                      "markersize":"3"})
```

### Hexbin plot

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
df
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Greys')
```

### KDE plot

```python
df2.plot.kde()
```
