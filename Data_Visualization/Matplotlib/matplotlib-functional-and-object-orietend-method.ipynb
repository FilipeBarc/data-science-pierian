```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Part 1

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
y
```

### Functional Method 

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### Scatter plot

```python
plt.scatter(x,y)
```

### Histogram

```python
from random import sample
data = sample(range(1,1000),100)
plt.hist(data,edgecolor='black')
```

### Boxplot

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

#rectangular boxplot
plt.boxplot(data,vert=True,patch_artist=True)
```

### Object Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig1 = plt.figure()

axes1 = fig1.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig1.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Lager Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

## Part 2

### Automating axes creation based on cols and rows - iteranting through the object

```python
fig,axes = plt.subplots(nrows=1,ncols=4)

for ax in axes:
    ax.plot(x,y)
    
plt.tight_layout()
```

### Indexing the object

```python
fig,axes = plt.subplots(nrows=1,ncols=4)

axes[1].plot(x,y)
axes[3].plot(y,x)

plt.tight_layout()
```

### Figure Size, aspect ratio and DPI

```python
fig = plt.figure(figsize=(8,2),dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2),dpi=100)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

### Creating legend

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')

ax.legend(loc=4)
```

### Saving plot

```python
fig.savefig('my_plot.jpg')
```

## Part 3

### Plot appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,0.5,0.5])

ax.plot(x,y,
        color='#6AFFFF',       # line color
        lw=2,                  # line width
        alpha=0.8,             # line transparence 
        ls='--',               # line shape
        marker='o',            # line marker
        markerfacecolor='red'  # line marker color 
       
       ) # RGB Hex Code
```
