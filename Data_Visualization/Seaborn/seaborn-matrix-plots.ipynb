```python
import seaborn as sns
import scipy
%matplotlib inline
```

### Built-in data set

```python
tip = sns.load_dataset('tips')
tip
```

```python
fly = sns.load_dataset('flights') 
fly
```

### Creating a correlational matrix

```python
tc = tip.corr()
tc
```

### Heat map

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

### Creating a pivot table

```python
fp = fly.pivot_table(index='month',columns='year',values='passengers')
fp
```

### Heat map

```python
sns.heatmap(fp,cmap='cool')
```

### Cluster map

```python
sns.clustermap(fp,cmap='coolwarm')
```
